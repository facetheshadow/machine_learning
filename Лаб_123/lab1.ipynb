```python
import os
import tarfile
import urllib.request
import pandas as pd
import plotly.graph_objects as go
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns
```

```python
data_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/lab1_dataset.csv")
print(data_df.to_string())
```

```python
# Group data by year and sum up the statistics
grouped_data = data_df.groupby('Year').sum()

# Plot the statistics by year
grouped_data.plot(kind='line', title='Crime Statistics by Year')

# Add axis labels
plt.xlabel('Year')
plt.ylabel('Number of Crimes')·

# Display the plot
plt.show()
```

```python
# Extract the data for 2010
data_2010 = data_df[data_df['Year'] == 2010]

# Calculate the total number of crimes for each region in 2010
region_totals = data_2010.groupby(['Unit Name'])['Total Cases'].sum()

# Calculate the proportion of crimes for each region in 2010
region_proportions = region_totals / region_totals.sum()

# Create a fan chart of the proportions
fig, ax = plt.subplots()
ax.pie(region_proportions, labels=region_proportions.index, startangle=90, counterclock=False, wedgeprops={'width': 0.4})
ax.set_title('Proportion of Crimes by Region in 2010')
plt.show()

# 设置绘图风格
plt.style.use('seaborn')

# 绘制直方图
plt.hist(region_proportions, bins=10, edgecolor='black')

# 添加标题和坐标轴标签
plt.title('Proportions of Crimes in Each Region in 2010')
plt.xlabel('Proportions')
plt.ylabel('Frequency')

# 显示图形
plt.show()








```

```python
# Define the data
labels = region_proportions.index
sizes = region_proportions.values

# Define the plot settings
fig, ax = plt.subplots()
ax.set_title('Proportion of Crimes by Region in 2010')

# Create the pie chart
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

# Set aspect ratio to be equal so that pie is drawn as a circle
ax.axis('equal')

# Show the plot
plt.show()
```

```python
# Data
regions = ['DMP']
years = ['2010', '2019']
crimes = ['Dacoity Robbery', 'Murder', 'Speedy', 'Vehicle Theft']

count_2010 = [123, 256, 178, 392]
count_2019 = [53, 86, 26, 135]

# Set the width of the bars
barWidth = 0.35

# Set position of bar on X axis
r1 = np.arange(len(count_2010))
r2 = [x + barWidth for x in r1]

# Plot the bars
plt.bar(r1, count_2010, color='blue', width=barWidth, edgecolor='grey', label=years[0])
plt.bar(r2, count_2019, color='orange', width=barWidth, edgecolor='grey', label=years[1])

# Add xticks on the middle of the group bars
plt.xticks([r + barWidth/2 for r in range(len(count_2010))], crimes)

# Set plot title and labels
plt.title('Number of Crimes in DMP Regions in 2010 and 2019')
plt.xlabel('Types of Crimes')
plt.ylabel('Number of Crimes')

# Add legend
plt.legend()

# Show plot
plt.show()
```
